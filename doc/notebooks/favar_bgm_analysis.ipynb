```python
%matplotlib inline
from __future__ import division

import time
import tables as tb
import numpy as np
import pandas as pd
import favar_bgm as fb
from tabulate import tabulate

from dismalpy import ssm
from dismalpy.ssm import dynamic_factors as df
from dismalpy.stats.rvs import gamma, normal, wishart
from dismalpy.reg import var
from dismalpy.reg._var import _var_posteriors, _var_quantities

import matplotlib.pyplot as plt
#import matplotlib.dates as mdates
np.set_printoptions(precision=10, suppress=True, linewidth=120)

columns = pd.read_csv('data/favar_bgm_columns.csv')
columns.index = columns.column
del columns['column']
columns.sort(inplace=True)
dates = pd.date_range('1976-02','2005-06', freq='MS')

# Get the data
raw = pd.read_csv('data/data76.csv', header=None)
raw.columns = columns.id.tolist()
raw.index = dates

# Only take the BBE dataset
#raw = raw.iloc[:,:110]

# Standardize the variables
dta = (raw - raw.mean()) / raw.std()

# Separate into observed economic variables and background variables
observed = dta['FYFF']
informational = dta.drop('FYFF', axis=1)
```

```python
# Get storage objects
start = 4500
with tb.open_file('favar_bgm.h5') as h5:
    table = h5.root.iterations.data
    
    data_states = table.read(start=start, field='states')
    data_obs_cov = table.read(start=start, field='obs_cov')
    data_loadings = table.read(start=start, field='loadings')
    data_state_cov = table.read(start=start, field='state_cov')
    data_phi = table.read(start=start, field='phi')
```

```python
mod = df.FAVAR(observed, informational, k_factors=5, order=13)
params = np.r_[
    data_loadings.mean(0).ravel(), data_obs_cov.mean(0),
    data_phi.mean(0), data_state_cov.mean(0).ravel()
]
mod.update(params)
mod.filter_method = ssm.FILTER_UNIVARIATE | ssm.FILTER_COLLAPSED
mod._initialize_representation()
mod._statespace.subset_design = True
mod._statespace.companion_transition = True
res = mod.smooth()
# states_median = np.median(data_states, 0)
smoothed_median = res.smoothed_state
```

### Impulse Response at Mean

```python
# Get VAR matrices
lamda = mod['design', :, :mod.k_posdef]
phi = mod['transition', :mod.k_posdef, :]
state_cov = mod['state_cov']

# LDL decomposition of the variance / covariance matrix
L = np.linalg.cholesky(state_cov)
D = np.diag(L.diagonal()**2)
L[np.diag_indices(L.shape[0])] = 1
```

```python
# Generate impulse response functions

# Dimensions
nirf = 128
nlags = 13
neqs = 6

# Storage
irf_tilde = np.zeros((nirf+nlags+1, neqs, neqs))
irf = np.zeros((nirf, neqs, neqs))

# Initial values
irf_tilde[nlags] = np.eye(neqs)*0.07
irf_tilde[nlags] = np.zeros((neqs,neqs))
irf_tilde[nlags][-1,-1] = 0.07
irf[0] = L

# Iteratively create IRFs
for i in range(nirf):
    for j in range(nlags):
        irf_tilde[i+nlags+1] += np.dot(phi[:,j*neqs:(j+1)*neqs], irf_tilde[i+nlags-j])
    irf[i] = np.dot(irf_tilde[i+nlags], L)
```

```python
# Impulse responses due to monetary policy shock
fig, axes = plt.subplots(neqs//2, 2, figsize=(14,10))

for i in range(neqs):
    axes[i%(neqs//2),i//(neqs//2)].plot(irf[:,i,5])
```

```python
# Extract the monetary IRFs
ffr_irfs = irf[:,:,5]

# Construct the loading matrix IRFs
info_irfs = np.zeros((nirf, mod.k_info))
for t in range(nirf):
    info_irfs[t] = np.dot(lamda[:-1,:], ffr_irfs[t])
```

### Median Impulse Response

```python
# Get VAR matrices

# LDL decomposition of the variance / covariance matrix
L = np.linalg.cholesky(data_state_cov)
diag = L.diagonal(axis1=1,axis2=2)**2
D = np.c_[[np.diag(diag[t]) for t in range(diag.shape[0])]]
idx = np.diag_indices(L.shape[1])
L[:, idx[0], idx[1]] = 1
```

```python
# Generate impulse response functions

# Dimensions
ndraws = L.shape[0]
nirf = 128
nlags = 13
neqs = 6

# Storage
irf_tilde = np.zeros((ndraws, nirf+nlags+1, neqs, neqs))
irf = np.zeros((ndraws, nirf, neqs, neqs))
info_irfs = np.zeros((ndraws, nirf, mod.k_info))

# Initial values
irf_tilde[:, nlags] = np.eye(neqs)*0.07
irf_tilde[:, nlags] = np.zeros((neqs,neqs))
irf_tilde[:, nlags, -1, -1] = 0.07
irf[:, 0] = L

# Iteratively create IRFs
for t in range(ndraws):
    # Get the parameter matrices
    params = np.r_[
        data_loadings[t].ravel(), data_obs_cov[t],
        data_phi[t], data_state_cov[t].ravel()
    ]
    mod.update(params)
    lamda = mod['design', :, :mod.k_posdef]
    phi = mod['transition', :mod.k_posdef, :]
    
    # Get the factor IRFs
    for i in range(nirf):
        for j in range(nlags):
            irf_tilde[t, i+nlags+1] += np.dot(phi[:,j*neqs:(j+1)*neqs], irf_tilde[t, i+nlags-j])
        irf[t, i] = np.dot(irf_tilde[t, i+nlags], L[t])
    
    # Extract the monetary IRFs
    ffr_irf = irf[t,:,:,5]

    # Construct the loading matrix IRFs

    for i in range(nirf):
        info_irfs[t,i] = np.dot(lamda[:-1,:], ffr_irfs[i])
```

```python
#median_irfs = np.median(irf, 0)
conf_irfs = np.percentile(irf, [5,50,95], 0)
#median_info_irfs = np.median(info_irfs, 0)
conf_info_irfs = np.percentile(info_irfs, [5,50,95], 0)
```
