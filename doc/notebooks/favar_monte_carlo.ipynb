```python
#%matplotlib inline
from __future__ import division

import time
import numpy as np
import pandas as pd
import favar_bbe as fb
from dismalpy import ssm
from dismalpy.ssm import dynamic_factors as df
from dismalpy.stats.rvs import gamma, normal, wishart

#import matplotlib.pyplot as plt
#import matplotlib.dates as mdates
np.set_printoptions(precision=5, suppress=True, linewidth=120)
```

```python
np.random.seed(1234)
# Generated dataset:

# 2 Factors + FFR
# VAR(3) => 6*3 = 18 parameters
# 6 (co)variance parameters
# 10 informational variables + 1 observed

# => (11-3) * 3 = 24 loading parameters
# 11 variance parameters

# Example parameters
nobs = 350
k_info = 10
k_obs = 1
k_endog = k_info + k_obs
order = 13
k_posdef = 3
k_states = 3*order
k_ar = order * k_posdef
k_var = k_posdef * k_ar

# There are two options:
# 1. Draw a "true" loading matrix relating all states
#    to informational variables, but which cannot
#    be estimated exactly due to the identification
#    strategy of setting the top K x K block of the
#    estimated loading matrix be the identity matrix
# loading = np.c_[
#     np.random.normal(size=(k_info,3)).T,
#     [0,0,1]
# ].T
# 2. Draw a "fake" loading matrix, enforcing the
#    K x K identity matrix.
true_loading = np.zeros((k_endog, k_posdef))
true_loading[:k_posdef-1,:k_posdef-1] = np.eye(k_posdef-1)
true_loading[k_posdef-1:-1,:] = np.random.normal(size=(k_info-(k_posdef-1), k_posdef))
true_loading[-1,-1] = 1

true_R = np.r_[np.random.chisquare(2, size=k_info), 1e-4] / 100

true_phi = np.random.normal(size=(k_posdef,k_states)) / 50
true_phi[np.diag_indices(k_posdef)] += 0.5
true_Q = next(wishart.Wishart(k_posdef, np.eye(k_posdef)/10))

# Generate example states
state_errors = normal.Normal(loc=[0,0,0], scale=true_Q)
true_states = np.zeros((k_posdef, nobs+order))
etas = np.zeros((k_posdef, nobs))
for t in range(order, nobs+order):
    lagged = true_states[:k_posdef,t-order:t][:,::-1].T.ravel()
    etas[:, t-order] = next(state_errors)
    eta = etas[:, t-order]
    for i in range(k_posdef):
        true_states[i,t] = np.dot(true_phi[i,:], lagged) + eta[i]
true_states_all = true_states[:,order:].T
for i in range(1,order):
    true_states_all = np.c_[
        true_states_all,
        true_states[:,order-i:-i].T
    ]
true_states = true_states[:,order:]

# Generate example observed (FFR) series
# with a small amount of measurement error
observed = (
    true_states[-1,:] +
    np.random.normal(scale=true_R[-1], size=(1,nobs))
).T

# Generate example informational series
informational = np.zeros((k_info, nobs))
for t in range(nobs):
    for i in range(k_info):
        informational[i,:] = (
            np.dot(true_states.T, true_loading[i]) +
            np.random.normal(scale=true_R[i], size=(nobs))
        )
informational = informational.T
```

```python
# Make sure the transition from the generated phi is stationary
transition = np.zeros((k_states,k_states))
transition[:k_posdef,:] = true_phi
transition[k_posdef:,:-k_posdef] = np.eye(k_states-k_posdef)
print np.max(np.abs(np.linalg.eigvals(transition)))
```

```python
# Define the Gibbs-Sampling iterations
def sample(model, G1, G):
    # Create the samplers
    sampler_state = model.simulation_smoother(simulation_output=ssm.SIMULATION_STATE)
    sampler_state._simulation_smoother.simulated_model.subset_design = True
    sampler_state._simulation_smoother.simulated_model.companion_transition = True
    
    sampler_obs_prec = gamma.Gamma(prior_obs_prec_shape / 2., 2. / prior_obs_prec_rate, preload=G1+G)
    sampler_loadings = normal.Normal(loc=prior_loading_mean, scale=prior_loading_var, preload=G1+G)
    sampler_state_prec = wishart.Wishart(prior_state_prec_df, prior_state_prec_scale, preload=G1+G)
    sampler_phi = normal.Normal(loc=prior_state_mean, scale=prior_state_var, preload=G1+G)
    sampler_phi._use_posterior = True
    
    # Create a VAR
    var_phi = var.VAR(nobs=model.nobs, k_endog=model.k_posdef, order=model.order)
    
    # Storage
    states = np.zeros((G+G1+1, model.k_states, model.nobs))
    obs_prec = np.zeros((G+G1+1, model.k_info))
    obs_cov = np.zeros((G+G1+1, model.k_info))
    loadings = np.zeros((G+G1+1, model.k_info-model.k_factors, model.k_posdef))
    state_prec = np.zeros((G+G1+1, model.k_posdef, model.k_posdef))
    state_cov = np.zeros((G+G1+1, model.k_posdef, model.k_posdef))
    phi = np.zeros((G+G1+1, model.k_var))
    
    # Initial values
    start_params = model.start_params
    loadings[0] = np.reshape(
        start_params[model._params_loadings],
        (model.k_info - model.k_factors, model.k_posdef)
    )
    obs_cov[0] = start_params[model._params_obs_cov]
    obs_prec[0] = 1. / obs_cov[0]
    phi[0] = start_params[model._params_transition]
    state_cov[0] = np.reshape(
        start_params[model._params_state_cov],
        (model.k_posdef, model.k_posdef)
    )
    state_prec[0] = np.linalg.inv(state_cov[0])

    # Run the iterations
    for i in range(G1+G):
        # Unobserved states
        states[i+1] = fb.sample_states(loadings[i], obs_cov[i], phi[i], state_cov[i], model, sampler_state)
        #states[i+1] = true_states_all.T
        
        # Observation equation (OLS)
        obs_prec[i+1] = fb.sample_obs_prec(loadings[i], states[i+1], model, sampler_obs_prec)
        obs_cov[i+1] = 1. / obs_prec[i+1]
        #obs_cov[i+1] = true_R[:-1]
        #obs_prec[i+1] = 1. / true_R[:-1]
        loadings[i+1] = fb.sample_loadings(obs_prec[i+1], states[i+1], model, sampler_loadings)
        #loadings[i+1] = true_loading[k_posdef-1:-1,:]
        
        # Transition equation (VAR)
        state_prec[i+1] = fb.sample_state_prec(phi[i], states[i+1], model, sampler_state_prec)
        state_cov[i+1] = np.linalg.inv(state_prec[i+1])
        #state_prec[i+1] = np.linalg.inv(true_Q)
        #state_cov[i+1] = true_Q
        phi[i+1] = fb.sample_phi(state_prec[i+1], states[i+1], model, var_phi, sampler_phi)
        #phi[i+1] = true_phi.ravel()
   
    return states[1:], obs_cov[1:], loadings[1:], state_cov[1:], phi[1:]
```
