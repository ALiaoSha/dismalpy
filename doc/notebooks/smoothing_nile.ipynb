```python
%matplotlib inline
import numpy as np
import pandas as pd
from datetime import datetime
import statsmodels.api as sm
from dismalpy import ssm

import matplotlib.pyplot as plt
```

```python
class LocalLevel(ssm.Model):
    def __init__(self, endog, *args, **kwargs):
        # Model order
        k_states = k_posdef = 1

        # Initialize the statespace
        super(LocalLevel, self).__init__(
            endog, k_states=k_states, k_posdef=k_posdef, *args, **kwargs
        )

        # Initialize the matrices
        self.design = np.r_[1]
        self.transition = np.r_[1]
        self.selection = np.eye(k_states)

        # Initialize the state space model as approximately diffuse
        self.initialize_approximate_diffuse(1e7)
        self.loglikelihood_burn = 1

    _latex_names = ['$\\sigma_\\varepsilon^2$', '$\\sigma_\\eta^2$']
    _names = ['sigma2.measurement', 'sigma2.level']
    def _get_model_names(self, latex=False):
        return self._latex_names if latex else self._names

    @property
    def start_params(self):
        return np.r_[0.1, 0.1]

    def transform_params(self, unconstrained):
        return unconstrained**2

    def untransform_params(self, constrained):
        return constrained**0.5

    def update(self, params, *args, **kwargs):
        params = super(LocalLevel, self).update(params, *args, **kwargs)

        # Extract the parameters
        measurement_variance = params[0]
        level_variance = params[1]

        # Observation covariance
        obs_cov = self.obs_cov.real.astype(params.dtype)
        obs_cov[0] = measurement_variance
        self.obs_cov = obs_cov

        # State covariance
        state_cov = self.state_cov.real.astype(params.dtype)
        state_cov[0] = level_variance
        self.state_cov = state_cov
```

### ARIMA Example 1: Arima

```python
# Dataset
data = pd.read_table('data/Nile.dat', skiprows=1, header=None, names=['flow'])
data.index = pd.date_range(start='01-01-1871', end='01-01-1970', freq='AS')

# Fit the model
mod = LocalLevel(data['flow'])
mod.update(np.r_[15099.0,1469.1])
#res = mod.fit()
res = mod.smooth()
print res.summary()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,6))

# 2.1 (i) Scatterplot, filtered state, confidence intervals
axes[0,0].plot(data.index, data['flow'], 'k.')
axes[0,0].plot(data.index, res.filtered_state[0])
se = res.filtered_state_cov[0,0]**0.5
axes[0,0].plot(data.index, res.filtered_state[0] + 1.96*se, 'k-', alpha=0.2)
axes[0,0].plot(data.index, res.filtered_state[0] - 1.96*se, 'k-', alpha=0.2)

# 2.1 (ii) Filtered state variance
axes[0,1].plot(data.index, res.filtered_state_cov[0,0])

# 2.1 (iii) Prediction errors
axes[1,0].plot(data.index[1:], res.forecasts_error[0,1:])

# 2.1 (iv) Prediction variance
axes[1,1].plot(data.index[1:], res.forecasts_error_cov[0,0,1:])

[axes[i,j].set(xlim=('1865','1970')) for i in [0,1] for j in [0,1]]
fig.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,6))

# 2.2 (i) Scatterplot, smoothed state, confidence intervals
axes[0,0].plot(data.index, data['flow'], 'k.')
axes[0,0].plot(data.index, res.smoothed_state[0])
se = res.smoothed_state_cov[0]**0.5
axes[0,0].plot(data.index, (res.smoothed_state[0] + 1.96*se).T, 'k-', alpha=0.2)
axes[0,0].plot(data.index, (res.smoothed_state[0] - 1.96*se).T, 'k-', alpha=0.2)

# 2.2 (ii) Smoothed state variance
axes[0,1].plot(data.index, res.smoothed_state_cov[0,0])
axes[0,1].set(ylim=(2300,4100))

# 2.2 (iii) Smoothing cumulant errors
axes[1,0].plot(data.index, res.scaled_smoothed_estimator[0,1:])

# 2.2 (iv) Smoothign variance cumulant
axes[1,1].plot(data.index, res.scaled_smoothed_estimator_cov[0,0,1:])
axes[1,1].set(ylim=(5e-5,12e-5))

[axes[i,j].set(xlim=('1865','1970')) for i in [0,1] for j in [0,1]]
fig.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,6))

# 2.2 (i) Smoothed measurment disturbance
axes[0,0].plot(data.index, res.smoothed_measurement_disturbance[0])

# 2.2 (ii) Smoothed state variance
axes[0,1].plot(data.index, res.smoothed_measurement_disturbance_cov[0,0]**0.5)

# 2.2 (iii) Smoothing cumulant errors
axes[1,0].plot(data.index, res.smoothed_state_disturbance[0])

# 2.2 (iv) Smoothign variance cumulant
axes[1,1].plot(data.index, res.smoothed_state_disturbance_cov[0,0]**0.5)

[axes[i,j].set(xlim=('1865','1970')) for i in [0,1] for j in [0,1]]
fig.tight_layout()
```
