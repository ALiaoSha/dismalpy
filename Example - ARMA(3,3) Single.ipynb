```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import optimize, signal
import statespace as ss
np.set_printoptions(precision=2, suppress=True)
```

```python
# Model setup
np.random.seed(123)
nobs = 200

# True hyperparameters
sigma = 100**0.5
phi = np.array([1.8, -1.2, 0.34])
theta = np.array([0.3, 0.3, 0.25])
true = np.r_[theta, phi, sigma**2]

# Generate dataset
v = np.random.normal(0, sigma, size=(nobs,))
y = signal.lfilter(np.r_[1, theta], np.r_[1, -phi], v)

# Plot datasets
fig = plt.figure(figsize=(14,4))
ax = fig.add_subplot(121)
ax.plot(y)
ax.set(title='Observed', xlabel=r'$t$', ylabel=r'$y_t$');
```
